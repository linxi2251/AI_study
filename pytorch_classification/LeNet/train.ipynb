```python
import torch
from torch.utils.data import DataLoader
import torchvision
from torchvision import transforms
import numpy as np
```

```python
transform = transforms.Compose(
    [transforms.ToTensor(),
     transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))]
)
```

```python
train_dataset = torchvision.datasets.CIFAR10(
    root="../../data_set", train=True, download=True, transform=transform)
```

```python
train_loader = DataLoader(train_dataset, batch_size=32,
                          shuffle=True, num_workers=8)
```

```python
val_dataset = torchvision.datasets.CIFAR10(
    root="../../data_set", train=False, download=True, transform=transform)
```

```python
val_loader = DataLoader(val_dataset, batch_size=64,
                          shuffle=True, num_workers=8)
```

```python
val_data_iter = iter(val_loader)
val_image, val_label = val_data_iter.__next__()
```

```python
import matplotlib.pyplot as plt
```

```python
def image_show(img):
    img = img / 2 + 0.5  # un_normalize
    np_img = img.numpy()
    plt.imshow(np.transpose(np_img, (1, 2, 0)))
    plt.show()
```

```python
image_show(torchvision.utils.make_grid(val_image))
```
